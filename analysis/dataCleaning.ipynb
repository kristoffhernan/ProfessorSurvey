```python
import pandas as pd
import numpy as np
import random as rd
```

```python
# read in csv
df = pd.read_csv("../data/professorData.csv")
```

```python
# get the column names
df.columns.values
```

```python
# replacing NAN with NA so that I can see Null and non-null count
df = df.iloc[:,:].replace("NAN",np.nan)
df.info()
```

```python
# look through all rows and cols for NAs
df[df.isnull().any(axis=1)].head()
```

```python
# keeping by last, bc biomedical sciences isnt too popular
# also Daniel R. Jeske is provost now, not chancellor
df[df.duplicated('FullName',keep='last')].sort_values(by='FullName').head()
df = df[~df.duplicated('FullName',keep='last')].sort_values(by='FullName')
```

```python

#[df.loc[index,'College'] == 'School of Medicine' for index in df.index if df.loc[index,'College'] == 'Biomedical Sciences']
df.loc[:,'College'].replace('Biomedical Sciences',"School of Medicine",inplace=True)

# double check to see if biomedical sciences is still there
df[df['College']=="Biomedical Sciences"]
```

```python
df[df['Title'].str.lower().str.contains("emeritus|emerita|emeriti")].head()
```

```python
# emeritus means retired 
df = df[~df['Title'].str.lower().str.contains("emeritus|emerita|emeriti")]
df[df.isnull().any(axis=1)]
```

```python
deceased = ['Fred Z Havens','William Junkert Jr']
new_practice = ['Steven Wilson MD','Uziel Sauceda','Walter M Marcus']
no_info = ['Y Paul Aoyagi']

filters = deceased + new_practice + no_info
df[df['FullName'].str.contains("|".join(filters))]
```

```python
df = df[~df['FullName'].str.contains("|".join(filters))]
```

```python
# double check for na's
df[df.isnull().any(axis=1)]
```

```python
mydict = {"Assistant": "Assistant Professor", "Associate": "Associate Professor", "Distinguished": "Distinguished Professor", "Dean": "Dean", "Chair": "Chair", "Lecturer":"Lecturer"}

# df = df.drop(columns='TitleNew')

for key, value in mydict.items():
    to_change = df['Title'].str.contains(key)
    df.loc[to_change,'TitleNew'] = value

df = df.fillna('Professor')
df = df.drop(columns='Title')

df.head()
```

```python
df.groupby(['TitleNew']).nunique()
# only 10 adjunct professors, so remove that filter
```

```python
# only 7 deans so have to remove 
df = df.loc[df['TitleNew'] != 'Dean']
df.head()
```

```python
df.groupby(['College']).nunique()
```

```python
# daniel jeske doesnt teach. the provost vice chancellor 
df = df.loc[~df['College'].str.contains("Provost")]
```

```python
# turn CE - College of Nat to College of Nat and Agr Sciences
df['College'].loc[df['College'] == 'CE - College of Nat & Ag Sci'] = 'Coll of Nat & Agr Sciences'

# double check
df.loc[df['College'] == 'CE - College']
```
